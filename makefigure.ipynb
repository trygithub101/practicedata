```
with codecs.open("basicinformation.csv", "r", "Shift-JIS", "ignore") as file:
    df = pd.read_table(file, delimiter=",", header =None)

    df = df.iloc[503:512, 1:23]
    df = df.drop(columns=[2,3,4,5,6,7,8])
print(df)

def plottv(i):
    
    xrange = []
    yrange = []

    for s, d in zip(list(df.columns)[1:],list(df.iloc[i,:])[1:]):#列colum from 1 to last, i行すべて from 1 to last
        xrange.append(float(s)-8)
        yrange.append(float(d))

    plt.plot(xrange, yrange, label = df.iloc[i,0])


for p in range(9):
    plottv(p)
 
plt.title("半沢直樹のドラマなど2013年度の視聴率推移")    
plt.xlabel("各話")
plt.ylabel("視聴率[%]")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=11)    
plt.show()
```
